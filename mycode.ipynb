```python
keep = df.drop_duplicates()
keep
```

```python
columns_to_keep = [ 'Cts', 'Col','Clarity','Cut','Pol', 'Sym', 'Fls']
orignal_keep=keep
keep = keep[columns_to_keep]
keep
```

```python
orignal_keep
```

```python
unique_df=keep.drop_duplicates()
unique_df
```

```python
# count the number of occurrences of each unique row in keep DataFrame
counts = keep.groupby(list(keep.columns)).size().reset_index(name='Count')
# merge counts back to unique_df
unique_df = unique_df.merge(counts, on=list(keep.columns), how='left')
```

```python
#dataframe showing all unique diamond configurations
unique_df = unique_df.sort_values(by='Count', ascending=False)
unique_df
```

```python
# function will going to return a dataframe which will include all the occurences of same diamond config
def show_Count_details(i):
    top_unique_df=unique_df.iloc[i]
    mask = (orignal_keep.Cts == top_unique_df.Cts) & (orignal_keep.Col == top_unique_df.Col) & (orignal_keep.Clarity == top_unique_df.Clarity) & (orignal_keep.Cut == top_unique_df.Cut) & (orignal_keep.Pol == top_unique_df.Pol) & (orignal_keep.Sym == top_unique_df.Sym) & (orignal_keep.Fls == top_unique_df.Fls) #& (orignal_keep.Amt == top_unique_df.Amt)
    return orignal_keep[mask]
```

```python
res=show_Count_details(0)
columns_to_keep = ['Company','Amt', 'Cts', 'Col','Clarity','Cut','Pol', 'Sym', 'Fls']
res=res[columns_to_keep]
res=res.sort_values(by='Company', ascending=False)
res
```

```python
# no of companies which produces given config diamonds
res['Company'].nunique()
# Average price of specified dimond produced by each company who make them.
res[['Company','Amt']].groupby(['Company']).mean()
```
